```python
import numpy as np
import maxpy as mp
from pythonosc.udp_client import SimpleUDPClient

from pythonosc.dispatcher import Dispatcher
from typing import List, Any
```

```python
#getting osc message from processing 
#when PLAYING: get number of stars + build maxpatch, then open maxpatch

dispatcher = Dispatcher()


def build_voices(address: str, *args: List[Any]) -> None:

    if not len(args) == 1 or type(args[0]) is not int:
        print("format")
        return

    # Check that address starts with filter
    if not address == "/numStars":
        print("add")
        return

    num_stars = args[0]


    yag = mp.MaxPatch(load_file = "yag-max-template.maxpat")

    route = yag.objs["obj-4"]
    voice = yag.objs["obj-3"]
    udp = yag.objs["obj-5"]
    out1 = yag.objs["obj-2"]
    out2 = yag.objs["obj-1"]

    i = 1
    while i != num_stars:
        nroute, nvoice = yag.place(["route /obj-" + str(i), "voice"])

        yag.connect((udp.outs[0], nroute.ins[0]),
                    (nroute.outs[0], nvoice.ins[0]),
                    (nvoice.outs[0], out1.ins[0]), 
                    (nvoice.outs[1], out2.ins[0]))

        #for debug
        nmsg = yag.place(["message " + str(i)])[0]
        yag.connect((nvoice.outs[1], nmsg.ins[0]),
                    (nmsg.outs[0], out2.ins[0]))
        #end debug

        i+= 1
        
    msg = yag.place(["message 0"])[0]
    yag.connect((voice.outs[1], msg.ins[0]),
                    (msg.outs[0], out2.ins[0]))
    
    
    yag.save("yag-voices.maxpat")
    



dispatcher.map("/numStars", build_voices)  # Map wildcard address to set_filter function

# Set up server and client for testing
from pythonosc.osc_server import BlockingOSCUDPServer
from pythonosc.udp_client import SimpleUDPClient

server = BlockingOSCUDPServer(("127.0.0.1", 3000), dispatcher)

# Send message and receive exactly one message (blocking)
server.serve_forever()
```
